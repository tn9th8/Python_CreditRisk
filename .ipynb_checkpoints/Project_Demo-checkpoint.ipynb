### Nhận xét:

* Nếu ta chuyển Age thành biến phân loại liệu có tốt hơn ?

```python
"""Đưa Age thành thành biến phân loại để phân tích"""
df['Age group'] = np.nan #Thêm cột, kiểu là nan "Not a Number"

df.loc[(df['Age'] >= 18) & (df['Age'] < 23), 'Age group'] = 'student'
df.loc[(df['Age'] >= 23) & (df['Age'] < 30), 'Age group'] = 'young'
df.loc[(df['Age'] >= 30) & (df['Age'] < 40), 'Age group'] = 'u30'
df.loc[(df['Age'] >= 40) & (df['Age'] < 50), 'Age group'] = 'u40'
df.loc[(df['Age'] >= 50) & (df['Age'] < 60), 'Age group'] = 'u50'
df.loc[(df['Age'] >= 60                   ), 'Age group'] = 'elder' 
    
df.head()

#loc: location, phương thức selection của pandas
```

```python
"""
Function: Tính tổng và phần trăm của bad, good loans theo age group
Paramters: 
    col1 là cột, cụ thể là Age group, 
    filter1 là giá trị cần lọc
Returns: 
    amount_all là tổng lượng tiền vay
    amount_good, amount_bad là tổng lượng tiền vay của good, bad credit
    percent_good, percent_bad là phần trăm của good, bad credit amount
"""
def compute_Credit_Amount(col1, filter1):
    amount_good = df['Credit amount'].loc[(df[col1] == filter1) & (df['Risk'] == 'good')].sum()
    amount_bad  = df['Credit amount'].loc[(df[col1] == filter1) & (df['Risk'] == 'bad')].sum()
    amount_all  = amount_good + amount_bad
    
    percent_good = amount_good/amount_all * 100
    percent_bad  = amount_bad /amount_all * 100
    # làm tròn số float, rồi chuyển sang string
    percent_good = str(round(percent_good, 2))
    percent_bad  = str(round(percent_bad , 2))
    
    return  amount_all, amount_good, amount_bad, percent_good, percent_bad

student_all, student_good, student_bad, student_good_p, student_bad_p = compute_Credit_Amount("Age group", "student")
young_all  , young_good  , young_bad  , young_good_p  , young_bad_p   = compute_Credit_Amount("Age group", "young")
u30_all    , u30_good    , u30_bad    , u30_good_p    , u30_bad_p     = compute_Credit_Amount("Age group", "u30")
u40_all    , u40_good    , u40_bad    , u40_good_p    , u40_bad_p     = compute_Credit_Amount("Age group", "u40")
u50_all    , u50_good    , u50_bad    , u50_good_p    , u50_bad_p     = compute_Credit_Amount("Age group", "u50")
elder_all  , elder_good  , elder_bad  , elder_good_p  , elder_bad_p   = compute_Credit_Amount("Age group", "elder")


x = df["Age group"].value_counts().index.values; print(x)
y1 = df["Age group"].value_counts().values # số lượng
y2 = [u30_all, young_all, u40_all, u50_all, student_all,  elder_all] # phần trăm
```

```python
# Tạo subplot với 1 hàng và 2 cột
fig = make_subplots(
    rows=1, cols=2, 
    subplot_titles=('Group age - Loan number', 'Group age - Credit amount'), 
    specs=[[{'type':'domain'}, {'type':'domain'}]])

# Tạo biểu đồ pie đầu tiên
fig1 = px.pie(
    df, 
    values=y1, 
    names=x)
fig.add_trace(fig1.data[0], 1, 1)

# Tạo biểu đồ pie thứ hai
fig2 = px.pie(
    df, 
    values=y2, 
    names=x)
fig.add_trace(fig2.data[0], 1, 2)

# Cập nhật vị trí và thông tin của text
fig.update_traces(textposition='inside', textinfo='percent+label')

fig.show()
```

# Explanations:

* Nhóm Young, U30, U40 ở lứa tuổi từ lúc vừa ra tốt nghiệp và <=49 tuổi.
    - Ở lứa tuổi này người ta có nhiều hoạch định và thích hiện thực chúng như (khởi nghiệp, mua nhà, mua xe, xây dựng gia đình, học cao học,...). Có lẽ vì thế mà họ là nhóm người cần nhiều tiền và là nhóm đi vay nhiều nhất
    - Tuy nhiên nhóm U30, U40 có sức khỏe, kiến thức, kinh nghiệm và khả năng lao động tốt nhất. Vì thế khả năng trả nợ của họ cũng tốt nhất
    - Còn nhóm Young lại rất thiếu kinh nghiệm, quyết định thì rất bồng bột. Vì thế rủi rõ thu hồi nợ ở họ là thấp, chỉ có ~50%
    
* Nhóm U50, Elder ở lứa tuổi >=50 tuổi.
    - Nhóm U50, Elder dường như cuộc sống đã ổn định, không còn nhiều ước mơ nên số người là nhóm này đi vay rất ít. 
    - Song, chính họ cũng đã người độ tuổi lao động. Có lẽ vì nguyên nhân này mà khiến khoản vay của U50 có rủi ro cao
    - Cái này của Elder lại có rủi ro thấp hơn, chắc vì họ có lương hưu
* Nhóm Student <= 22 tuổi
    - Nhóm Student còn trẻ, có nhiều ước mơ. Thay vì hiện thực hóa, họ tập trung cho học tập, tích lũy kinh nghiệm. Bổ sung thêm, chi phí học tập đại học ở Đức có thể hoàn toàn chi trả nhờ vào việc đi làm thêm. Vì thế mà nhóm này có lại rất ít người đi vay.
    - Và đặc biệt, rủi ro ở nhóm này cũng thấp. Tuy thiếu nhiều kinh thức và kinh nghiệm, nhưng ở Đức, những khoản vay của họ lại nhận nhiều ưu đãi dành cho sinh viên với lại suất thấp, hoặc đó có thể là khoản vay hỗ trợ học tập của Chính phủ. Tiềm năng trả nợ cũng lớn vì tương lai có thể đi làm

* Unskilled, non-resident: 
    - Hầu hết là những người thất nghiệp và lao động nước ngoài nên việc có thể mượn nợ là rất khó

* Skilled: 
    - Có kỹ năng chuyên môn, nghề nghiệp ổn định --> Rất ít khi vay và cũng có khả năng thanh toán nợ do nguồn thu nhập linh hoạt nên tỉ lệ nợ xấu thấp --> Lịch sử thanh toán tốt --> Thường là good risk

### Nghề nghiệp và số tiền trong thẻ tín dụng của họ?
* Khách hàng có tín dụng tốt thường có xu hướng có số tiền tín dụng cao hơn khách hàng có tín dụng xấu. Có được điều này là do Thu nhập, tài sản thế chấp theo từng nhóm nghề nghiệp. Do đó, các ngân hàng sẵn sàng cho vay nhiều hơn, nguy cơ vỡ nợ thấp hơn-->Good risk
* Khách hàng có tín dụng tốt có phân phối số tiền tín dụng hẹp hơn khách hàng có tín dụng xấu. Cho thấy KH có tín dụng tốt có số tiền tín dụng ít biến động hơn do có thu nhập cao hơn, có kế hoạch tài chính tốt hơn, có ý thức trả nợ cao hơn
* Ngoài ra có một số điểm ngoại lệ trong biểu đồ hộp cho KH có tín dụng tốt. Do thu nhập hoặc tài sản bất thường cao như trúng xổ số, thừa kế hoặc do công việc mà được nhận 1 khoản tiền lớn. Lịch sử tín dụng tốt nhưng có khoản nợ lớn gần đây điều này tạo ra điều bất thường trên biểu đồ nhưng vẫn được xem là tín dụng tốt

#### Mục địch vay dựa trên độ tuổi
    Rất đa dạng nhưng người lớn tuổi có khả năng vay mượn vì mục đích có lợi nhiều hơn người ít tuổi(do kinh nghiệm) nhưng chia theo nhóm "Good risk" và "Bad risk" thì gồm một số nguyên nhân sau. *Good risk*: Nhóm này khoảng sau 30 tuổi có thể bao gồm những người ổn định về mặt tài chính, đã hoàn thành giai đoạn mục tiêu tài chính lớn, và có khả năng chi trả các khoản tín dụng một cách đáng tin cậy. Họ vay để phục vụ cho công việc kinh doanh hoặc dành cho giáo dục bản thân, con cái,... *Bad risk*: Nhóm này khoảng 20 đến 40 có thể chứa những người đang trong giai đoạn cuộc sống khó khăn, có nhiều nghĩa vụ tài chính, và gặp khó khăn trong việc thanh toán các khoản vay. Họ phải vay để có thể chi trả cho các khoảng chi tiêu trong cuộc sống điều này dẫn đến nhiều rủi ro xấu

    *Nhu Cầu Tài Chính Lớn*: Có thể mục đích sử dụng đó liên quan đến các nhu cầu tài chính lớn, như mua nhà, ô tô, hay đầu tư kinh doanh, đòi hỏi số tiền vay lớn hơn.
    Người Vay Mượn Có Thu Nhập Cao: Những người chọn mục đích sử dụng này có thể có thu nhập cao, giúp họ đáp ứng được các khoản vay lớn.
    
    *Nhu Cầu Tài Chính Nhỏ*: Các mục đích sử dụng này có thể liên quan đến nhu cầu tài chính nhỏ hơn, như du lịch, giáo dục, hay mua sắm hàng ngày, không đòi hỏi mức vay lớn.
    Người Vay Mượn Có Thu Nhập Thấp Hơn: Những người chọn các mục đích sử dụng này có thể có thu nhập thấp, có khả năng chi trả các khoản vay nhỏ hơn.

```python
"""import ML library"""
from sklearn.model_selection import train_test_split, KFold, cross_val_score # to split the data
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, fbeta_score #To evaluate our model

from sklearn.model_selection import GridSearchCV

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
#from sklearn.xgboost import XGBClassifier

df_ml = df.copy()
df_ml.head(2)
```

```python
"""xóa những cột có null value (NA)""" #Note: this cell only run 1 time. Because of using a delete method
# tìm columns có giá trị null
sr0 = df_ml.isnull().sum().sort_values(ascending=False)
col0 = np.array([])
for id, val in sr0.iteritems(): #lấy từng ele của 1 series pandas =))
    if val > 0:
        col0 = np.append(col0, id)

#delete null column
for val in col0:
    df_ml = df_ml.drop(val, axis=1)
    
df_ml.head(2)
```

```python
"""lấy unique values của biến category"""
cates = dict() #dict category: {each column name: list unique values} 
cols = np.array(df_nb.columns)
for col in cols:
    cates.update({col: df_nb[col].unique()}) #or: cates[col] = df_nb[col].unique()

for ele in cates.items():
    if ele[0] not in ["Age", "Credit amount", "Duration"]:
        print(ele)
```

```python
"""chuyển unique values của biến category thành numeric"""
def transform_data_to_numeric(column):
    i = 0
    for val in cates[column]: # cates là một dict()
        df_ml.loc[df[column] == val, column] = i
        i += 1
transform_data_to_numeric("Sex") 
transform_data_to_numeric("Housing")
transform_data_to_numeric("Purpose") 

df_ml.head(2)
```

```python
"""chia data thành 2 tập training và testing"""
X = df_ml.drop('Risk', 1).values #.values: cover frame => array, 1 record => 1 element
y = df_ml["Risk"].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state=42)

X_train[5]
```

```python
"""so sánh ML algorithms for classification"""
models = [] #prepare models
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('RF', RandomForestClassifier()))
models.append(('SVM', SVC(gamma='auto')))
#models.append(('XGB', XGBClassifier()))

#đánh giá từng model
results = []
names = []
kfold = KFold(n_splits=10, random_state=7, shuffle=True)
for name, model in models:
        cv_results = cross_val_score(model, X_train, y_train, cv=kfold)
        results.append(cv_results)
        names.append(name)
        msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
        print(msg)
        
# boxplot algorithm comparison
fig = plt.figure(figsize=(11,6))
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```
